Direction le Suuuuuud üòé

Sur cette page, tu vas trouver le panorama des festivals en PACA. T√©l√©charge le fichier Json et extrais ses donn√©es.
Dis-nous combien il y a de festivals en PACA, quelles sont les 3 villes qui accueillent la majorit√© des festivals et quel est le mois de l'ann√©e le plus charg√© en festivals.

```python
# Load the JSON file
df_festivals = pd.read_json("/content/panorama-des-festivals.json")

# Display the first few rows of the dataframe
df_festivals.head()
```

Il semble que le nom de la r√©gion n'est pas disponible directement dans les donn√©es. Cependant, nous avons le d√©partement ('dept_sk') et nous savons que la r√©gion PACA comprend les d√©partements suivants :

04 : Alpes-de-Haute-Provence

05 : Hautes-Alpes

06 : Alpes-Maritimes

13 : Bouches-du-Rh√¥ne

83 : Var

84 : Vaucluse

Je vais donc utiliser ces informations pour filtrer les festivals qui se d√©roulent en PACA. Ensuite, j'identifierai les trois villes qui accueillent le plus de festivals, et je d√©terminerai le mois le plus charg√© en festivals.

```python
# Extract information from the 'fields' column and create a new dataframe
df_festivals_details = pd.json_normalize(df_festivals['fields'])

# Define the department codes for PACA region
paca_dept_codes = [4, 5, 6, 13, 83, 84]

# Filter festivals in PACA
df_festivals_paca = df_festivals_details[df_festivals_details['dept_sk'].isin(paca_dept_codes)]


# Display the column names of the dataframe with festival details
df_festivals_details.columns
```

Il semble que le nom de la commune est stock√© dans la colonne 'libelle_commune_pour_calcul_cp_insee'. Je vais utiliser cette colonne pour identifier les trois villes qui accueillent le plus de festivals. Ensuite, je d√©terminerai le mois le plus charg√© en festivals.

```python
# Count the number of festivals in PACA
num_festivals_paca = df_festivals_paca.shape[0]

# Identify the top 3 cities hosting the most festivals
top_cities = df_festivals_paca['libelle_commune_pour_calcul_cp_insee'].value_counts().nlargest(3)

# Find the busiest month for festivals
busiest_month = df_festivals_paca['mois_indicatif_en_chiffre_y_compris_double_mois'].value_counts().idxmax()

num_festivals_paca, top_cities, busiest_month
```

Il y a 319 festivals dans la r√©gion PACA (Provence-Alpes-C√¥te d'Azur).

Les trois villes qui accueillent le plus grand nombre de festivals sont :

Marseille 01 avec 56 festivals
Cannes avec 17 festivals

Donner le nombre de clients par ville, class√© par ordre alphab√©tique de villes.


```python
# Reconnect to the SQLite database
conn = sqlite3.connect("/content/py066-ma_banque.db")

# Execute a SQL query to get the number of clients per city, ordered alphabetically by city
client_counts_per_city = pd.read_sql_query("""
SELECT ville, COUNT(*) as nombre_de_clients FROM client
GROUP BY ville
ORDER BY ville
""", conn)

# Close the connection
conn.close()

# Display the client counts per city
client_counts_per_city
```

Il semble que Paris soit la ville avec le plus grand nombre de clients (6 clients), suivie de Lyon, Marseille et Strasbourg (5 clients chacune).

Donner la ville ayant le plus de clients.


```python
# Reconnect to the SQLite database
conn = sqlite3.connect("/content/py066-ma_banque.db")

# Execute a SQL query to get the city with the most clients
city_with_most_clients = pd.read_sql_query("""
SELECT ville, COUNT(*) as nombre_de_clients FROM client
GROUP BY ville
ORDER BY nombre_de_clients DESC
LIMIT 1
""", conn)

# Close the connection
conn.close()

# Display the city with the most clients
city_with_most_clients
```

La ville qui a le plus de clients est Paris, avec 6 clients.

Trouver le nombre d‚Äôop√©rations effectu√©es sur chaque compte.

```python
# Reconnect to the SQLite database
conn = sqlite3.connect("/content/py066-ma_banque.db")

# Execute a SQL query to get the number of operations per account
operation_counts_per_account = pd.read_sql_query("""
SELECT idcompte, COUNT(*) as nombre_d_operations FROM operation
GROUP BY idcompte
""", conn)

# Close the connection
conn.close()

# Display the operation counts per account
operation_counts_per_account
```

Il y a donc 288 comptes en tout, avec un nombre d'op√©rations allant de 2 √† 8 par compte.
